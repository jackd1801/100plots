```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
## Import data (2021 edition) - from https://www.ons.gov.uk/peoplepopulationandcommunity/personalandhouseholdfinances/incomeandwealth/datasets/householddisposableincomeandinequality
```

```python
filename = "Data/ONSincomeData.xlsx"

#Import dataframe, make data long and clean up variable types

df = pd.read_excel(filename, 'Table 3', skiprows = 6, nrows=45,  usecols = 'B:K')
df = df.tail(-1)
df['Year'] = df.Year.astype(str).str[:4]
df = pd.melt(df, id_vars=['Year'], var_name='Decile', value_name='Income')
df['Decile'] = df.Decile.str.extract('(\d+)')
df['Decile'] = pd.to_numeric(df['Decile'])
df['Income']=pd.to_numeric(df['Income'])

#create data for first decile 

first = df.copy()
first = first.loc[first["Decile"] == 2]
first['Decile']=1
df = pd.concat([df,first])
df = df.sort_values(by=['Decile', 'Year'])

#Create total and shares 

Total=df.groupby('Year')['Income'].sum()
df = pd.merge(df,Total,on='Year',how='outer')
df = df.rename(columns={"Income_x": "Income", "Income_y": "Total"})
df['Share']=df['Income']/df['Total']

#Plotting dataframe test

plotdf = pd.DataFrame().assign(Year=df['Year'], Share=df['Share'], Decile=df['Decile'])
plotdf = pd.pivot(plotdf, index=['Year'], columns = 'Decile',values = 'Share')
```

```python
plot = plotdf.plot.area()
```

```python
filename = "Data/ONSincomeData.xlsx"

#Import dataframe, make data long and clean up variable types

df = pd.read_excel(filename, 'Table 2', skiprows = 8, nrows=45,  usecols = 'B:G')
df = df.tail(-3)
df['Year'] = df.Year.astype(str).str[:4]
df.columns = df.columns.str.strip() 
df = df.rename(columns={"Bottom": 1, "2nd": 2, "3rd": 3, "4th":4, "Top": 5})
df = pd.melt(df, id_vars=['Year'], var_name='Quintile', value_name='Income')

Total=df.groupby('Year')['Income'].sum()
df = pd.merge(df,Total,on='Year',how='outer')
df = df.rename(columns={"Income_x": "Income", "Income_y": "Total"})
df['Share']=df['Income']/df['Total']

plotdf = pd.DataFrame().assign(Year=df['Year'], Share=df['Share'], Quintile=df['Quintile'])
plotdf = pd.pivot(plotdf, index=['Year'], columns = 'Quintile', values = 'Share')
```

```python
plot = plotdf.plot.area()
```
